```python

lakehouse_silver_path=spark.conf.get('spark.executorEnv.lakehouse_silver')
#lakehouse_gold_path=spark.conf.get('spark.executorEnv.lakehouse_gold')
lakehouse_silver_table=spark.conf.get('spark.executorEnv.lakehouse_silverTable')
#lakehouse_gold_table=spark.conf.get('spark.spark.executorEnv.lakehouse_goldTable')

print(lakehouse_silver_path)
#print(lakehouse_gold_path)
print(lakehouse_silver_table)
```

```python
payers_df = spark.read.format("csv").option("header","true").load("Files/raw/payers.csv")
# df now is a Spark DataFrame containing CSV data from "Files/raw/payers.csv".
display(payers_df)
```

```python
payers_transformed=payers_df.dropna(subset=['id'])
payers_transformed.count()
```

```python
payers_cleaned=payers_transformed.withColumnRenamed('id','DIM_payerId')
```

```python
payers_clean=payers_cleaned.dropDuplicates()
```

```python
payers_clean.printSchema()
```

```python
payers_clean.write.format('delta').mode('overwrite').save(f"{lakehouse_silver_table}/DIM_Payers")
```
