```python

lakehouse_silver_table=spark.conf.get('spark.executorEnv.lakehouse_silverTable')

print(lakehouse_silver_table)
```

```python
organizations_df = spark.read.format("csv").option("header","true").load("Files/raw/organizations.csv")
# df now is a Spark DataFrame containing CSV data from "Files/raw/organizations.csv".
display(organizations_df)
```

```python
org_df=organizations_df.dropna()
```

```python
org_cleaned=org_df.dropDuplicates()
```

```python
org_clean=org_cleaned.withColumnRenamed('id','DIM_organizationId')
```

```python
org_clean.printSchema()
```

```python
org_clean.write.format('delta').mode('overwrite').save(f"{lakehouse_silver_table}/DIM_Organizations")
```
